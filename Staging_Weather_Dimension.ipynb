```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
ottawa_w= pd.read_csv("en_climate_daily_ON_6105976_2020_P1D.csv")
```

```python
ottawa_w.head()
```

```python
weather_dimension = pd.DataFrame(columns=['weather_surrogate_key','daily_high', 'daily_low', 'rain_amount',
                                     'snow_amount', 'date_col','station_name'])
```

```python
counter = 0 
start_date = datetime(2020, 9, 1)
end_date = datetime(2020, 12, 31)

for idx, row in ottawa_w.iterrows():
    curr_date = row['Date/Time']
    curr_date = datetime.strptime(curr_date,'%Y-%m-%d')
    if(curr_date >= start_date and curr_date <= end_date):
        #Weather
        weather_row = [idx + 100]
        weather_row.append(row['Max Temp (°C)'])
        weather_row.append(row['Min Temp (°C)'])
        weather_row.append(row['Total Precip (mm)'])
        weather_row.append(row['Total Snow (cm)']) 
        weather_row.append(curr_date)
        weather_row.append(row['Station Name'][:6].strip())
        weather_dimension.loc[len(weather_dimension)] = weather_row
```

```python
toronto_w = pd.read_csv("en_climate_daily_ON_6158731_2020_P1D.csv")
```

```python
toronto_w.head()
```

```python
start_date = datetime(2020, 9, 1)
end_date = datetime(2020, 12, 31)

for idx, row in toronto_w.iterrows():
    curr_date = row['Date/Time']
    curr_date = datetime.strptime(curr_date,'%Y-%m-%d')
    if(curr_date >= start_date and curr_date <= end_date):
        #Weather
        weather_row = [idx+300]
        weather_row.append(row['Max Temp (°C)'])
        weather_row.append(row['Min Temp (°C)'])
        weather_row.append(row['Total Precip (mm)'])
        weather_row.append(row['Total Snow (cm)']) 
        weather_row.append(curr_date)
        weather_row.append(row['Station Name'][:7].strip())
        weather_dimension.loc[len(weather_dimension)] = weather_row
```

```python
weather_dimension.fillna(0, inplace = True)
```

```python
weather_dimension.to_csv("weather_dimension.csv",index=False)
```
