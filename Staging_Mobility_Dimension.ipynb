```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
mobility_data= pd.read_csv("2020_CA_Region_Mobility_Report.csv")
```

```python
mobility_data.head()
```

```python
mobility_dimension = pd.DataFrame(columns=['mobility_surrogate_key','metro_area', 'subregion', 'province',
                                           'retail_and_recreation','grocery_and_pharmacy','parks','transit_stations','workplaces','residential','date_col'])
```

```python
mobility_dimension.head()
```

```python
start_date = datetime(2020, 9, 1)
end_date = datetime(2020, 12, 31)
counter = 0
for idx, row in mobility_data.iterrows():
    counter = counter + 1
    curr_date = row['date']
    subregion = row['sub_region_2']
    curr_date = datetime.strptime(curr_date,'%Y-%m-%d')
    if(curr_date >= start_date and curr_date <= end_date):
        if(subregion == "Regional Municipality of Durham" or subregion == "Regional Municipality of Halton" 
          or subregion == "Regional Municipality of Peel" or subregion == "Regional Municipality of York"
          or subregion =="Toronto Division" or subregion == "Ottawa Division"):
            #Mobility
            mobility_row = [counter]
            mobility_row.append(row['metro_area'])
            mobility_row.append(row['sub_region_2'])
            mobility_row.append(row['sub_region_1'])
            mobility_row.append(row['retail_and_recreation_percent_change_from_baseline']) 
            mobility_row.append(row['grocery_and_pharmacy_percent_change_from_baseline']) 
            mobility_row.append(row['parks_percent_change_from_baseline']) 
            mobility_row.append(row['transit_stations_percent_change_from_baseline']) 
            mobility_row.append(row['workplaces_percent_change_from_baseline']) 
            mobility_row.append(row['residential_percent_change_from_baseline']) 
            mobility_row.append(curr_date)
            mobility_dimension.loc[len(mobility_dimension)] = mobility_row
```

```python
mobility_dimension.head()
```

```python
mobility_dimension.to_csv("mobility_dimension.csv",index=False)
```
